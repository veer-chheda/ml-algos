```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

```python
y = df['logS']
y
```

```python
x = df.drop('logS',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
lr  = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_r2_train = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_r2_test = r2_score(y_test,y_lr_test_pred)
```

```python
print(lr_train_mse)
print(lr_r2_train)
print(lr_test_mse)
print(lr_r2_test)
```

```python
lr_results = pd.DataFrame(['Linear Regression',lr_train_mse,lr_r2_train,lr_test_mse,lr_r2_test]).transpose()
lr_results.columns = ['Method','Train MSE','Train R2','Test MSE','Test R2']
lr_results
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(5,5))
plt.scatter(x=y_train,y=y_lr_train_pred,alpha=0.4)
z = np.polyfit(y_train,y_lr_train_pred,1)
p = np.poly1d(z)
plt.plot(y_train, p(y_train),'#000000')
plt.xlabel('Experimental logS')
plt.ylabel('logS Prediction')
```

```python
from sklearn.linear_model import Ridge
r = Ridge(alpha=0.001)
r.fit(X_train,y_train)
```

```python
r_y_train_pred = r.predict(X_train)
print('MSE train',mean_squared_error(y_train,r_y_train_pred))
print('R2 Score train',r2_score(y_train,r_y_train_pred))
```

```python
r_y_test_pred = r.predict(X_test)
print('MSE test',mean_squared_error(y_test,r_y_test_pred))
print('R2 Score test',r2_score(y_test,r_y_test_pred))
```
